```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
```

```python
G = nx.karate_club_graph()
columns = [f'Node {i}' for i in range(0, G.number_of_nodes())]
degrees = [G.degree(v) for v in G]

dicDF = {
    columns[i] : degrees[i] for i in range(len(degrees))
}

degreesV = pd.DataFrame({
    'Nodes' : columns, 
    'Degrees' : degrees
    }
)

degreesV.set_index(['Nodes'], inplace=True)
degreesV
```

```python
plt.figure(figsize=(15, 8))
plt.title("Zachary's Karate Club Network")
nx.draw_circular(G, with_labels=True, node_color='orange')
plt.savefig("Timelapse/Zachary's_Karate_Club_Network", dpi=300)
plt.show()
```
